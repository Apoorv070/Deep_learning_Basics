```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##  Loading the data

```python
X_train = pd.read_csv("Logistic_X_Train.csv").values
```

```python
Y_train = pd.read_csv("Logistic_Y_Train.csv").values
```

```python
X_test = pd.read_csv("Logistic_X_Test.csv").values
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
plt.style.use("seaborn")
plt.scatter(X_train[:,0],X_train[:,1],c=Y_train,cmap=plt.cm.Accent)
plt.show()
```

```python
from keras import models
from keras.layers import Dense
```

```python
model = models.Sequential()
model.add(Dense(16,activation='relu',input_shape = (2,)))
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
#model compile
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
```

```python

model.summary()
```

```python
X_val = X_train[:1125]
X_train_new = X_train[1125:]

Y_val = Y_train[:1125]
Y_train_new = Y_train[1125:]
```

```python
hist  = model.fit(X_train_new,Y_train_new,epochs=700,batch_size=512,validation_data=(X_val,Y_val))
```

```python
h = hist.history
```

```python
plt.plot(h['val_loss'],label = "Validation Loss")
plt.plot(h['loss'],label="Training Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
output = model.predict(X_test)
```

```python
m = output.shape[0]
```

```python
for i in range(m):
    if(output[i]>0.5):
        output[i]=int(1)
    else:
        output[i]=int(0)
```
